```python
import pandas as pd

#bowlers = pd.read_excel("ODI_cricket.xlsx", sheet_name="bowler")
#display(bowlers.head(10))
```

#Creating DataFrame with our lists

## Education Data

```python
df_A = pd.DataFrame()
df_A["Name"] = ["Sama", "Sabab", "Samira"]
df_A["Education"] = ["Graduate", "College", "Undergraduate"]
df_A["Subject"] = ["CSE", "Maths", "BBA"]

display(df_A)
```

## Names Data

```python
df_B = pd.DataFrame() 
df_B["Name"] = ["Sama", "Samira", "Sabab"]
df_B["Last Name"] = ["Samrin", "Sabah", "Salam"]
df_B["Nickname"] = ["Orpa", "Suha", "Sabab"]

display(df_B)
```

## Personal Info Data

```python
df_C = pd.DataFrame()
df_C["Name"] = ["Sama", "Shabiba", "Sabab", "Samira"]
df_C["Location"] = ["Uttara", "Banani", "Dhanmondi", "Khilkhet"]
df_C["Age"] = ["27", "23", "20", "21"]
df_C["Education"] = ["Graduate", "Graduate", "College", "Undergraduate"]

display(df_C)
```



---



# Joining Data

## Inner Join

```python
#Joining based on Names
df_names = df_A.merge(df_B, how="inner", on=["Name"])

display(df_names)
```

```python
df_education = df_C.merge(df_A, how="inner", on=["Education"])

display(df_education)
```

```python
df_education_name = df_A.merge(df_C, how="inner", on=["Education", "Name"])

display(df_education_name)
```

## Left Join

```python
df_left = df_A.merge(df_C, how="left", on=["Name"])

display(df_left)
```

## Right Join

```python
df_right = df_A.merge(df_C, how="right", on=["Name"])

display(df_right)
```

## Outer Join

```python
df_outer = df_A.merge(df_C, how="outer", on=["Name"])

display(df_outer)
```
